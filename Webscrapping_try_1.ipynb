```python
#Import Beautiful Soup to help in breaking down the web page content
from bs4 import BeautifulSoup
import requests
```

```python
#Define the web that will be extract
hidden_figures = 'https://en.wikipedia.org/wiki/Hidden_Figures'
```

```python
#Create HTTP GET request to the targeted web
response = requests.get(hidden_figures)
```

```python
#Store the HTML content in a variable
html_content = response.text
```

```python
#Parse the html content from the targeted website
hiddenfigures_soup = BeautifulSoup(html_content, 'html.parser')
```

```python
# Display a snippet of the HTML content
print(html_content[:500])
```

**Navigating the HTML Structure:**

```python
# Find all <a> tags (anchor tags) in the HTML
links = hiddenfigures_soup.find_all('a')

# Iterate through the list of links and print their text
for link in links:
    print(link.text)
```
